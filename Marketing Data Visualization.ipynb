```python
data_cat = data[['Education', 'Marital_Status', 'Kidhome', 'Teenhome', 'AcceptedCmp3',
                 'AcceptedCmp1', 'AcceptedCmp2', 'Response', 'Complain', 'Country']]
for i in data_cat.columns:
    plt.figure(figsize=(10,5))
    sns.countplot(data_cat[i], data = data_cat, palette = 'hls')
    plt.xticks(rotation = 90)
    plt.show()
```

```python
for i in data_cat.columns:
    plt.figure(figsize=(10,5))
    data_cat[i].value_counts().plot(kind = 'pie', autopct = '%1.1f%%')
    plt.xticks(rotation = 90)
    plt.show()
```

```python
data_num = data[['ID', 'Year_Birth','Kidhome', 'Teenhome', 'Recency', 'MntWines','MntFruits','MntMeatProducts',
                 'MntFishProducts', 'MntSweetProducts','MntGoldProds', 'NumDealsPurchases','NumCatalogPurchases',
                 'NumStorePurchases', 'NumWebVisitsMonth','AcceptedCmp4','AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2',
                 'AcceptedCmp4','Response', 'Complain']]
```

```python
for i in data_num.columns:
    plt.figure(figsize=(5,1))
    sns.boxplot(data_num[i])
    plt.xticks(rotation = 90)
    plt.show()
```
